```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
a1=stats.norm.rvs(loc=9.4,scale=0.5,size=70)
a2=stats.norm.rvs(loc=12.0,scale=0.7,size=70)
a3=stats.norm.rvs(loc=10.3,scale=0.6,size=70)
df=pd.DataFrame({
    "A1":a1,
    "A2":a2,
    "A3":a3
})
sns.violinplot(data=df,jitter=True)
plt.hlines(xmin=-0.6,xmax=2.6,y=df.mean().mean(),color="red",linestyles="dashed")
plt.xlabel("factor",size=14)
# plt.ylabel("value",size=14)
plt.savefig("violinplot")
```
