### 確率pをモデル化
### pのロジット変換

```python
def p(x,b0,b1):
    return 1.0/(sp.e**(-(b0+b1*x))+1)
```

```python
x=np.arange(-8,8+1,0.1)
plt.title("logistic distribution",size=15)
plt.plot(x,p(x,-1,1))
```

```python
plt.plot(x,p(x,0,1.5))
```
