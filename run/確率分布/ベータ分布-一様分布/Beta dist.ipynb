### ベータ分布のPDFとCDF

```python
def beta_pdf(x,alpha,beta):
    return stats.beta.pdf(x,alpha,beta)
def beta_cdf(x,alpha,beta):
    return stats.beta.cdf(x,alpha,beta)
```

```python
x = np.arange(0,1.01,0.01)
alpha,beta = 30,70
y = beta_pdf(x,alpha,beta)
cdf=beta_cdf(x,alpha,beta)
```

```python
plt.plot(x,y)
plt.plot(x,cdf,color="orange")
print("mean:",alpha/(alpha+beta))
print(" var:",(alpha*beta)/((alpha+beta+1)*(alpha+beta)**2))
```
