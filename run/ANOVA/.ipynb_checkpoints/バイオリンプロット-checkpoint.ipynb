```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
a1=stats.norm.rvs(loc=1,scale=1,size=20)
a2=stats.norm.rvs(loc=-0.3,scale=2,size=20)
a3=stats.norm.rvs(loc=1.2,scale=1.5,size=20)
```

```python
df=pd.DataFrame({
    "A1":a1,
    "A2":a2,
    "A3":a3
})
```

```python
sns.violinplot(df)
```
