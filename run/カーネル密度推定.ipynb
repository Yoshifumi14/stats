```python
import numpy as np
import scipy as sp
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import math
sns.set()
```

### カーネル関数(ガウス関数;平均0,分散1)    

```python
def k(x):
    return stats.norm.pdf(x=x,loc=0,scale=sp.sqrt(1))
#     return (1.0/sp.sqrt(2*math.pi))*sp.exp((-x**2)/2)
```

#### 正規分布のPDF(ガウス関数)

```python
tmp_x=np.arange(-10,10)
plt.plot(k(tmp_x))
```

## カーネル密度推定量

```python
def kde(a_,x,h):
    a=np.array(a_)
    return (1.0/(len(a)*h))*sp.sum(k((x-a)/h))
#     return (1/(len(a)*h))*sp.sum(stats.norm.pdf(k((x-a)/h),loc=0,scale=sp.sqrt(0.5)))
```

### 実験

```python
a=np.array([-2.1,-1.3,-0.4,1.9,5.1,6.2])
# k_=[]
x_list=np.arange(-5,10)
# k_=kde(a,x_list,1)
k_=[]
for x in x_list:
    k_.append(kde(a,x,1))
plt.plot(x_list,k_)
plt.hist(a,bins=15)
```
